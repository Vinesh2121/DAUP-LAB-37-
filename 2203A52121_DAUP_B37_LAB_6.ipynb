```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
```

```python
# Load the dataset
df = pd.read_csv("/content/Bank-Customer-Attrition-Insights-Data.csv")
```

```python
# Display basic information
display(df.info())
display(df.head())
```

```python
df.columns = df.columns.str.lower()
expected_columns = ['balance', 'creditscore', 'transactioncount']
existing_columns = [col for col in expected_columns if col in df.columns]
if not existing_columns:
    raise KeyError("None of the expected numerical columns found in dataset")
if not existing_columns:
    raise KeyError("None of the expected numerical columns found in dataset")

if existing_columns:
    summary_stats_standard = df[existing_columns].agg(['mean', 'median', 'std'])

    summary_stats_custom = df[existing_columns].apply(lambda x: pd.Series({'skew': skew(x), 'kurtosis': kurtosis(x)}))

    summary_stats = pd.concat([summary_stats_standard, summary_stats_custom], axis=1)

    print("Summary Statistics:")
    print(summary_stats)
```

```python
categorical_columns = ['gender', 'geography', 'age']
categorical_columns = [col for col in categorical_columns if col in df.columns]
for col in categorical_columns:
    print(f"\nDistribution of {col}:")
    print(df[col].value_counts())
```

```python
if 'transactioncount' in df.columns:
    active_customers = df[df['transactioncount'] > 0]
    inactive_customers = df[df['transactioncount'] == 0]

    print("\nActive Customers Transaction Count Statistics:")
    print(active_customers['transactioncount'].agg(['min', 'max', 'mean']))

    print("\nInactive Customers Transaction Count Statistics:")
    print(inactive_customers['transactioncount'].agg(['min', 'max', 'mean']))
```

```python
if existing_columns:
    correlations = df[existing_columns].corr()
    print("\nCorrelation Matrix:")
    print(correlations)
```

```python
if 'churn' in df.columns:
    churned = df[df['churn'] == 1]
    stayed = df[df['churn'] == 0]

    churned_stats = churned[['balance', 'transactioncount']].describe()
    stayed_stats = stayed[['balance', 'transactioncount']].describe()

    print("\nStatistics for Churned Customers:")
    print(churned_stats)
    print("\nStatistics for Customers Who Stayed:")
    print(stayed_stats)
else:
    print("\n'Churn' column not found in dataset.")
```

```python
if 'creditscore' in df.columns and 'balance' in df.columns:
    low_credit_threshold = df['creditscore'].quantile(0.25)
    high_balance_threshold = df['balance'].quantile(0.75)

    total_customers = len(df)
    low_credit_high_balance_customers = df[(df['creditscore'] <= low_credit_threshold) & (df['balance'] >= high_balance_threshold)]
    probability = len(low_credit_high_balance_customers) / total_customers

    print(f"\nProbability of a customer having both low credit score and high balance: {probability:.4f}")
```
